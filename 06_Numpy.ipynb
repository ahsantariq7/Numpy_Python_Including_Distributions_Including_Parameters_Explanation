# Difference between copy and view in numpy_array

- copy--New array(make new error from existing array) so therefore now its value is defined in new array
- View--You can see and read although it makes it temporary copy but can't count in copy
- we can see that copy depends on new variable in which we copy the data of 1st variable.On the otherhand,in view, exist variable is responsible for any change .

#### we will make a copy, change in its origonal array,and display both.

```python
ahsan=np.array([1,2,3,4,5])
ahsan1=ahsan.copy()
ahsan1[0]=12
print(ahsan,ahsan1)
```

#### now we will make a view,change original, display both

```python
ahsan=np.array([1,2,3,4,5,6,7])
ahsan1=ahsan.view()
ahsan[0]=12
print(ahsan,ahsan1)
```

# Numpy Array Iterating

-  means going through the elements one by one or step by step like For loop

### Iterate the 1_D Array

```python
x=np.array([1,2,3,4,5])
for i in x:
    print(i,end=" ")
```

### Iterate the 2_D Array

```python
x=np.array([[1,2,3,4],[3,4,5,6]])
for i in x:
    print(i)
```

### Iterate on each scalar element of the 2_D

```python
for i in x:
    for j in i:
        print(j,end=" ")
```

### Iterate on 3_D Array

```python
a= np.array([[[1,2,3,4],[2,3,4,5],[4,5,6,7],[4,5,6,7]]])
for i in a:
    print(i)
```

### Iterate on each scalar element of 3_D

```python
for i in a:
    for j in i:
        for k in j:
            print(k,end=" ")
```

```python
# Python code to demonstrate the
# use of numpy.cov
import numpy as np

x = [1.23, 2.12, 3.34, 4.5]

y = [2.56, 2.89, 3.76, 3.95]

# find out covariance with respect columns
cov_mat = np.stack((x, y), axis = 0)

print(np.cov(cov_mat))
```

```python
# Python code to demonstrate the
# use of numpy.cov
import numpy as np

x = [1.23, 2.12, 3.34, 4.5]

y = [2.56, 2.89, 3.76, 3.95]

# find out covariance with respect rows
cov_mat = np.stack((x, y), axis = 1)

print("shape of matrix x and y:", np.shape(cov_mat))

print("shape of covariance matrix:", np.shape(np.cov(cov_mat)))

print(np.cov(cov_mat))
```
